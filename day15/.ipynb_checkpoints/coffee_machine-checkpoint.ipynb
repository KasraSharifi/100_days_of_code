```python
def coffee_machine_software():
    """This software can be used to monitor the management of the coffee machine"""
    from menu import MENU

    resources = {
        "water": 300,
        "milk": 200,
        "coffee": 100
    }
    
    profit = 0

    def calculate_the_order(ordered_drink):
        """THis function sums the value of all the inserted coins and checks to see if the money is enought to 
        order a drink. It also calculates the chnages if the inserted coins value is more than drink price."""

        print("please insert coins")
        quarters = int(input("how many quarters? "))
        dimes = int(input("how many dimes? "))
        nickels = int(input("how many nickles? "))
        pennies = int(input("how many pennies? "))
        total_money = (0.25 * quarters) + (0.1 * dimes) + (0.05 * nickels) + (0.01 * pennies)
        if total_money >= MENU[ordered_drink]["cost"]:
            change = total_money - MENU[ordered_drink]["cost"]
            print(f"Here is ${round(change, 2)} in change")
        else:
            print("not enough money. money refunded")
            print(f"you need to inster ${MENU[ordered_drink]['cost']} worth of coins.")
            calculate_the_order(ordered_drink = ordered_drink)


    def do_we_have_enough_resources(ordered_drink):
        """This functions checks the ingrediensts in the coffee machine to make sure that we have enough ingredients
        to make the ordered drink"""
        if MENU[ordered_drink]["ingredients"]["water"] > resources["water"]:                                
            print("Sorry. there is not enough water in the machine")
            return False
        elif MENU[ordered_drink]["ingredients"]["coffee"] > resources["coffee"]:
            print("Sorry. there is not enough coffee in the machine")
            return False
        if ordered_drink in ["latte", "cappuccino"]:
            if MENU[ordered_drink]["ingredients"]["milk"] > resources["milk"]:
                print("Sorry. there is not enough milk in the machine")
                return False


    while resources["water"] >= 50 and resources["coffee"] >= 18:
        action = input("click on 'Report' to get the resources or click on 'Order' to order a drink.")
        while action not in ["Report", "Order"]:
            print("the chosen action is not suported, please choose either Report or Drink")
            action = input("click on 'Report' to get the resources or click on 'Order' to order a drink.")
        else:
            if action == "Report":
                print(f"We have {resources['water']} water, {resources['milk']} milk, and {resources['coffee']} coffee")
                print(f"Also the totla profit is {profit}")
            else:    
                drink_type = input("What would you like? (espresso/ latte/ cappuccino)")
                while drink_type not in ["espresso", "latte", "cappuccino"]:
                    print(f"Unfortunately we cannot make {drink_type} drink.")
                    print(f"please choose either espresso, latte, or cappuccino")
                    drink_type = input("What would you like? (espresso/ latte/ cappuccino)")
                if drink_type == "espresso":
                    if do_we_have_enough_resources("espresso") == False:
                        print(":(")
                    else:
                        resources["water"] = resources["water"] - MENU["espresso"]["ingredients"]["water"]
                        resources["coffee"] = resources["coffee"] - MENU["espresso"]["ingredients"]["coffee"]
                        calculate_the_order("espresso")
                        print(f"Here your {drink_type}. ENJOY!!")
                        profit += MENU["espresso"]["cost"]
                elif drink_type == "latte":
                    if do_we_have_enough_resources("latte") == False:
                        print(":(")
                    else:
                        resources["water"] = resources["water"] - MENU["latte"]["ingredients"]["water"]
                        resources["coffee"] = resources["coffee"] - MENU["latte"]["ingredients"]["coffee"]
                        resources["milk"] = resources["milk"] - MENU["latte"]["ingredients"]["milk"]
                        calculate_the_order("latte")
                        print(f"Here your {drink_type}. ENJOY!!")
                        profit += MENU["latte"]["cost"]
                elif drink_type == "cappuccino":
                    if do_we_have_enough_resources("cappuccino") == False:
                        print(":(")
                    else:
                        resources["water"] = resources["water"] - MENU["cappuccino"]["ingredients"]["water"]
                        resources["coffee"] = resources["coffee"] - MENU["cappuccino"]["ingredients"]["coffee"]
                        resources["milk"] = resources["milk"] - MENU["cappuccino"]["ingredients"]["milk"]
                        calculate_the_order("cappuccino")
                        print(f"Here your {drink_type}. ENJOY!!")
                        profit += MENU["cappuccino"]["cost"]
```

```python
coffee_machine_software()
```
