```python
def play_higher_lower_game():
    import random
    from art import logo, vs
    from game_data import data
    print(logo)

    player_score = 0
    can_we_continue_play_the_game = True
    while can_we_continue_play_the_game == True:
        if player_score == 0:
            first_item, second_item = random.choices(data, k = 2)
        else:
            first_item = second_item
            second_item = random.choice(data)
        print(f"Compare A: {first_item['name']} a {first_item['description']} from {first_item['country']}")
        print(vs)
        print(f"Against B: {second_item['name']} a {second_item['description']} from {second_item['country']}")
        user_guess = input("Who has more followers? 'A' or 'B'? ")
        while user_guess not in ["A", "B"]:
            print("please choose either A or B")
            user_guess = input("Who has more followers? 'A' or 'B'? ")

        folower_count_a =  first_item["follower_count"]
        folower_count_b =  second_item["follower_count"]

        if folower_count_a > folower_count_b and user_guess == "A":
            player_score += 1
            print("==" * 30)
            print(f"You are right. Current score {player_score}")
            print("==" * 30)
        elif folower_count_a < folower_count_b and user_guess == "B":  
            player_score += 1
            print("==" * 30)
            print(f"You are right. Current score {player_score}")
            print("==" * 30)
        else:
            print("==" * 30)
            print(f"Sorry. That is wrong. Final score {player_score}")
            can_we_continue_play_the_game = False
            print("==" * 30)
            
    do_you_want_to_play_again = input("Would u like to play one more time? 'y' or 'n'")
    if do_you_want_to_play_again == "y":
        print("==" * 40)
        print("==" * 40)
        play_higher_lower_game()
```
