```python
def lets_play_black_jack():
    """
    We use this function to play a game of black jack
    """
    import random

    intention = input("Do you want to play a game of blac jack? type 'y' or 'n'.")

    while intention == "y":
        #define the whole deck
        card_deck = (list(range(2,12)) + [10,10,10] ) * 4
        
        #you start with having two cards
        user_cards = random.choices(card_deck, k = 2)
        print(f"your cards are {user_cards}")
        
        #now let's give dealers her cards
        dealer_cards = random.choices(card_deck, k = 2)
        print(f"dealer's first card {dealer_cards[0]}")
        
        want_one_more_card = input("Type 'y' to get another card, type 'n' to pass.")
        while want_one_more_card == "y":
            third_card = random.choice(card_deck)
            user_cards.append(third_card)
            print(f"now you have {user_cards} list of cards")
            if sum(user_cards) > 21:
                print("Your deck exceeded 21. You lost the game")
                want_one_more_card = "n"
                intention = input("Do you want to play another round? 'y'/'n'")
                print("==" * 30)
            else:
                want_one_more_card = input("Type 'y' to get another card, type 'n' to pass.")
        
        if sum(user_cards) > 21:
            continue
        print(f"dealer's hand {dealer_cards}")

        #if dealer's hand is less than 17, dealer must take another card 
        dealer_number_of_extra_card_drawn = 0
        while sum(dealer_cards) < 17:
            dealer_number_of_extra_card_drawn += 1
            print(f"dealer's hand total is less than 17, dealer has taken {dealer_number_of_extra_card_drawn} more cards till now")
            dealer_next_card = random.choice(card_deck)
            dealer_cards.append(dealer_next_card)
            print(f"dealer's final hand {dealer_cards}")
        
        if sum(dealer_cards) > 21:
            print("the sum of dealer's cards is more than 21. You won.")
        else:
            print(f"the sume of user's cards is {sum(user_cards)}")
            print(f"the sume of dealer's cards is {sum(dealer_cards)}")
            if sum(user_cards) > sum(dealer_cards):
                print("You won")
            elif sum(user_cards) == sum(dealer_cards):
                print("Draw")
            else:
                print("You lost the game")
        intention = input("Do you want to play another round? 'y'/'n'")
        print("==" * 30)
```
